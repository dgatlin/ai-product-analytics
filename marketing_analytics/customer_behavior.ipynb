# Exploratory Analysis for Customer Behavior

```python
import pandas as pd
import matplotlib.pyplot as plt

class ConsumerBehavior:
    def __init__(self, data_file):
        self.df = pd.read_csv(data_file)

    def overall_engagement_rate(self):
        response_counts = self.df.groupby('Response').count()['Customer']
        engagement_rate = response_counts / self.df.shape[0]
        return engagement_rate

    def plot_overall_engagement_rate(self):
        engagement_rate = self.overall_engagement_rate()
        ax = engagement_rate.plot(
            kind='bar',
            color='skyblue',
            grid=True,
            figsize=(10, 7),
            title='Marketing Engagement'
        )
        ax.set_xlabel('Engaged')
        ax.set_ylabel('Count')
        plt.show()

    def engagement_rate_by_offer_type(self):
        engaged_customers = self.df.loc[self.df['Response'] == 'Yes']
        by_offer_type = (
            engaged_customers.groupby('Renew Offer Type').count()['Customer'] /
            self.df.groupby('Renew Offer Type').count()['Customer']
        )
        return by_offer_type

    def plot_engagement_rate_by_offer_type(self):
        engagement_rate = self.engagement_rate_by_offer_type()
        ax = (engagement_rate * 100.0).plot(
            kind='bar',
            figsize=(7, 7),
            color='skyblue',
            grid=True
        )
        ax.set_ylabel('Engagement Rate (%)')
        plt.show()

    # Add more methods for other sections of the code...

# Example usage:
data_file = '../data_sets/Value-Analysis.csv'
consumer_behavior = ConsumerBehavior(data_file)

engagement_rate = consumer_behavior.overall_engagement_rate()
print("Overall Engagement Rate:")
print(engagement_rate)

consumer_behavior.plot_overall_engagement_rate()

engagement_rate_by_offer_type = consumer_behavior.engagement_rate_by_offer_type()
print("Engagement Rate by Offer Type:")
print(engagement_rate_by_offer_type)

consumer_behavior.plot_engagement_rate_by_offer_type()
```
